```python
from datetime import datetime
```

```python
class User:
    
    def __init__(self, name, email, age, height, alumn, birth):
        self.name = name
        self.email = email
        self.age = age
        self.height = height
        self.alumn = alumn
        self.birth = birth
        
    def __str__(self):
        return f"""
        Name: {self.name} 
        Email: {self.email}
        Age: {self.age}
        Height: {self.height}
        Alumn: {self.alumn}
        Birth: {self.birth}"""
        
  
```

```python
usuarios = [
    User("Cecilia Orosco", "corosco@gmail.com", 1.68, 54, True, "27/11/1970"),
    User("Manuel Cobos", "mcobos@gmail.com", 1.80, 58, False, "11/11/1966"),
    User("Ivan Cobos", "icobos@gmail.com", 1.82, 21, True, "07/04/2003")]
  
```

```python
def list_user():
    print("Has elegido la opción 1")
    if not usuarios: 
        print("No hay usuarios disponibles")
        return
    email2 = input("Introduce el email del usuario: ")
    be = False
    for usu in usuarios:
        if email2.lower() == usu.email.lower():
            print(usu)
            be = True
            break
        else:
            be = False
    if be == False:    
        print("usuario inexistente")
    
def list_all():
    print("Has elegido la opción 2")
    if not usuarios: 
        print("No hay usuarios disponibles")
        return
    for usu in usuarios:
        print(usu)
        print("\n")

def list_orden():
    print("Has elegido la opción 3")
    if not usuarios: 
        print("No hay usuarios disponibles")
        return
    while True:
        choice = int(input("""Ingrese:
                        1- Orden ascendente 
                        0- Orden descendente 
                        9- Salir"""))
        match choice:         
            case 1:
                user_sort_by_email = sorted(usuarios, key=lambda u : u.email)
                for us in user_sort_by_email:
                    print(us)
            case 0:
                user_sort_by_email = sorted(usuarios, key=lambda u : u.email, reverse=True)
                for us in user_sort_by_email:
                    print(us)
            case 9:
                break
            case _: 
                print("Opción incorrecta")
                
def crea_user():
    print("Has elegido la opción 4")
    print("Introduce los siguientes datos:")
    email2 = input("Email: ")
    be = False
    for usu in usuarios:
        if email2.lower() == usu.email.lower():
            print("Usuario existente")
            be = True
            break
    if be == False:    
        nom = input("Nombre: ")
        while True:
            try:
                edad = int(input("Edad: "))
                if edad >= 0:
                    break 
            except ValueError:
                print("Entrada no válida. Por favor, ingrese un número entero.")
        while True:
            try:
                altu = float(input("Altura: "))
                if altu > 0:
                    break 
            except ValueError:
                print("Entrada no válida. Por favor, ingrese un número con punto y al menos un decimal")
        while True:
            try:
                alum = bool(input("Curso (si = 1, no = 0)"))
                if alum == 1 or alum == 0:
                    break 
            except ValueError:
                print("Entrada no válida. Por favor, ingrese 1 ó 0")
        nac = solicitar_fecha()
        new_alumn = User(nom, email2, altu, edad, alum, nac)
        usuarios.append(new_alumn)
        print("Usuario creado correctamente")
        
def change_user():
    print("Has elegido la opción 5")
    if not usuarios: 
        print("No hay usuarios disponibles")
        return
    email2 = input("Introduce el email del usuario: ")
    be = False
    for usu in usuarios:
        if email2.lower() == usu.email.lower():
            print(usu)
            while True:
                choice = int(input(menu2))
                match choice:
                    case 1:
                        usu.name = input("Nombre: ")        
                    case 2:
                        usu.email = input("Email: ")
                    case 3:
                        while True:
                            try:
                                usu.age = int(input("Edad: "))
                                if usu.age >= 0:
                                    break 
                            except ValueError:
                                print("Entrada no válida. Por favor, ingrese un número entero.")
                    case 4:
                        while True:
                            try:
                                usu.height = float(input("Altura: "))
                                if usu.height > 0:
                                    break 
                            except ValueError:
                                print("Entrada no válida. Por favor, ingrese un número con punto y al menos un decimal")
                    case 5:
                        while True:
                            try:
                                usu.alumn = bool(input("Curso (si = 1, no = 0)"))
                                if usu.alumn == 1 | usu.alumn == 0:
                                    break 
                            except ValueError:
                                print("Entrada no válida. Por favor, ingrese 1 ó 0")
                    case 6:
                        usu.birth = solicitar_fecha()
                    case 7:
                        break
                    case _: 
                        print("Opción incorrecta")  
            be = True
            break
        else:
            be = False
    if be == False: 
        print("usuario inexistente")

def del_user():
    print("Has elegido la opción 6")
    if not usuarios: 
        print("No hay usuarios disponibles")
        return
    email2 = input("Introduce el email del usuario a borrar: ")
    be = False
    for usu in usuarios:
        if email2.lower() == usu.email.lower():
            usuarios.remove(usu)
            print("Usuario eliminado correctamente")
            be = True
            break
        else:
            be = False
    if be == False:
        print("usuario inexistente")
    
def del_all():
    print("Has elegido la opción 7")
    if not usuarios: 
        print("No hay usuarios disponibles")
        return
    usuarios.clear()
    #for i, x in enumerate(usuarios): otra manera de hacerlo
    #    del usuarios[i]
    print("Todos los usuarios eliminados")
    
def solicitar_fecha():
    while True:
        fecha_entrada = input("Fecha nacimiento (dd/mm/aaaa): ")
        fecha = validar_fecha(fecha_entrada)
        if fecha:
            return fecha
        else:
            print("Fecha no válida. Por favor, inténtelo de nuevo.")

def validar_fecha(fecha_str):
    try:
        fecha_valida = datetime.strptime(fecha_str, "%d/%m/%Y").date()
        return fecha_valida
    except ValueError:
        return None    
            
```

```python
menu = """Te damos la bienvenida a la app de Usuarios, elige una opción:
          1- Listar un usuario
          2- Listar todos los usuarios
          3- Listar los usuarios ordenados por email
          4- Crear un usuario
          5- Actualizar un usuario
          6- Borrar un usuario
          7- Borrar todos los usuarios
          8- Salir
         """

menu2 = """¿Qué dato quiere actualizar? Elige una opción: 
        1- Nombre
        2- Email
        3- Edad
        4- Altura
        5- Estudia
        6- Fecha nacimiento
        7- Guardar y salir
        """
```

```python
while True:
    prueba = int(input(menu))
    match prueba:
        case 1:
            list_user()
        case 2:
            list_all()
        case 3:
            list_orden()
        case 4:
            crea_user()
        case 5:
            change_user()
        case 6:
            del_user()
        case 7:
            del_all()
        case 8:
            print("Hasta pronto.")
            break
        case _: 
            print("Opción incorrecta")
```
