```python
#crear conexion a base de datos
connection = con.connect(
    host="localhost",
    port= "3306",
    user= "root",
    password= "admin"
)
```

```python
cursor = connection.cursor()
cursor.execute("SHOW DATABASES;")
# Procesar los resultados:
for name in cursor:
    print(name)
    
cursor.close()
connection.close()
```

```python
cursor = connection.cursor()
cursor.execute("SELECT * FROM sakila.customer;")
customers = cursor.fetchall()
for customer in customers:
    print(customer)
```

```python
cursor.execute("SELECT * FROM sakila.film;")

films = cursor.fetchall()
for film in films:
    print(film)
    
cursor.close()
```

```python
connection = con.connect(
    host="localhost",
    port= "3306",
    user= "root",
    password= "admin",
    database= "sakila" #poniendo aqui la base de datos luego no hace falta ponerlo en el execute()
)
cursor = connection.cursor()
cursor.execute("SELECT * FROM film;")

films = cursor.fetchall()
for film in films:
    print(film)
    
cursor.close()
```

```python
#sin poner la base de datos en la conexion
connection = con.connect(
    host="localhost",
    port= "3306",
    user= "root",
    password= "admin"#, 
    #database= "sakila"
)
cursor = connection.cursor()
#cursor.execute("SHOW TABLES;")
cursor.execute("SHOW TABLES from sakila;")


tables = cursor.fetchall()
for table in tables:
    print(table)
    
cursor.close()
```

```python
def sentencias(data, sentence):
    connection = con.connect(
        host="localhost",
        port= "3306",
        user= "root",
        password= "admin",
        database= data
    )
    cursor = connection.cursor()
    cursor.execute(sentence)
    tables = cursor.fetchall()
    for table in tables:
        print(table)
    cursor.close()
```

```python
sentencias("sakila", "SHOW TABLES")
```
